```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./Data/anaconda-2022-SODS-raw-data.csv', skiprows=2, header=1)
df
```

```python
# Create a dictionary so we can get the survey question by providing the code
questions_map_df = pd.read_csv('./Data/anaconda-2022-SODS-raw-data.csv', skiprows=2, nrows=1)
questions_map_df

code_to_question_map = {}
for (colName, colData) in questions_map_df.iteritems():
    code_to_question_map[str(colData.values[0])] = colName
    
```

```python
# Different roles in this dataset
unique_roles = df["1.6"].unique()
unique_roles
```

```python
# What proportion of this dataset are students? Get percentages per roles.
def get_count_by_role_name(role_name: str):
    selection = df[df["1.6"] == role_name]
    return selection.shape[0]
```

```python
data = []

total_rows = df.shape[0]
for role in unique_roles:
    role_count = get_count_by_role_name(role)
    data.append(role_count/total_rows)
    
roles_percentages_df = pd.DataFrame(unique_roles, columns=["Role"])
roles_percentages_df["Percentage"] = data
roles_percentages_df.sort_values(by=["Percentage"], ascending=False)
```

```python
# What questions were the most relevant to students?
students_df = df[df["1.6"] == "Student"]
total_students_count = students_df.shape[0]
# Total students
print("Total students: %s" % total_students_count)

data = []
pd.isnull(students_df["2"].values).sum()
for (colCode, colData) in students_df.iteritems():
    total_answered = total_students_count - pd.isnull(colData.values).sum()
    percentage = total_answered/total_students_count
    data.append((colCode, percentage))
    #columns.append(code_to_question_map[colName])
    
most_relevant_questions_df = pd.DataFrame(data, columns=["Code", "Percentage"])
filter = (most_relevant_questions_df["Percentage"] < 1) & (most_relevant_questions_df["Percentage"] > 0)
sorted_most_relevant_questions_df = most_relevant_questions_df[filter].sort_values(by=["Percentage"], ascending=False)

top_10_codes_df = sorted_most_relevant_questions_df[0:10]
top_10_codes = top_10_codes_df["Code"].values
top_10_percentages = top_10_codes_df["Percentage"].values
print("%s \n" % top_10_codes)
print("%s \n" % top_10_percentages)

for code in top_10_codes:
    print("%s  \n" % code_to_question_map[code])
    
```

```python
# Q5.1_1: These are Likert scale questions and also has a 'not applicable' option
print(code_to_question_map["5.1_1"])
print(students_df["5.1_1"].unique())
```

```python
non_null_answers = total_students_count - pd.isnull(students_df["5.1_1"]).sum()
num_frequently = students_df[students_df["5.1_1"] == "Frequently"].shape[0]
num_always = students_df[students_df["5.1_1"] == "Always"].shape[0]
num_sometimes = students_df[students_df["5.1_1"] == "Sometimes"].shape[0]
num_rarely = students_df[students_df["5.1_1"] == "Rarely"].shape[0]
num_never = students_df[students_df["5.1_1"] == "Never"].shape[0]

print("Student usage of Python language:")
data = [
    (num_always/non_null_answers, "Always"), 
    (num_frequently/non_null_answers, "Frequently"), 
    (num_sometimes/non_null_answers, "Sometimes"), 
    (num_rarely/non_null_answers, "Rarely"), 
    (num_never/non_null_answers, "Never")]
student_usage_of_python_df = pd.DataFrame(data, columns=["Percentage", "Frequency"])
student_usage_of_python_df
```

```python
# Q4.1: These are "select all that apply" questions
print(code_to_question_map["4.1"])

# Total answers
print("Total answers: %s" % (total_students_count - pd.isnull(students_df["4.1"].values).sum()))

# Iterate to break down into unique list
unique_selections_set = set()
all_selections_set = []

for item in students_df["4.1"]:
    
    # skip if is nan
    if pd.isnull(item):
        continue
        
    # filter because of comma interfering with split    
    if "Proprietary tools" in item:
        item = item.replace("Proprietary tools (ex. Tableau, Google Colab, SAS, MATLAB, etc)", "")
    
    # split on comma
    split_line = item.split(",")    
    
    # remove empties and strip whitespaces
    split_line = [l.strip() for l in split_line if l]
    
    # re-add filtered
    split_line.append("Proprietary tools (ex. Tableau, Google Colab, SAS, MATLAB, etc)")   
    
    for y in split_line:
        unique_selections_set.add(y)
        all_selections_set.append(y)
        
print("Unique Selections: %s" % unique_selections_set)
```

```python
# Analyze frequency of each topic/tool/skill - top ~5 and bottom ~5
data = []

for item in unique_selections_set:
    count = all_selections_set.count(item)
    data.append((count, item))

frequency_topics_tools_skills_df = pd.DataFrame(data, columns=["Frequency", "Label"])
frequency_topics_tools_skills_df.sort_values(by=["Frequency"], ascending=False)
```

```python
# Q4.4: These are "select all that apply" questions
print(code_to_question_map["4.4"])

# Total answers
print("Total answers: %s" % (total_students_count - pd.isnull(students_df["4.4"].values).sum()))
```

```python
# Analyze frequency
unique_selections_set = set()
all_selections_set = []

for item in students_df["4.4"]:    
    # skip if is nan
    if pd.isnull(item):
        continue
    
    # split on comma
    split_line = item.split(",")    
    
    # remove empties and strip whitespaces
    split_line = [l.strip() for l in split_line if l]
    
    for y in split_line:
        unique_selections_set.add(y)
        all_selections_set.append(y)

data = []

for item in unique_selections_set:
    count = all_selections_set.count(item)
    data.append((count, item))

frequency_topics_tools_skills_df = pd.DataFrame(data, columns=["Frequency", "Label"])
frequency_topics_tools_skills_df.sort_values(by=["Frequency"], ascending=False)
```

```python
# Q4.7
print(code_to_question_map["4.7"])

# Total answers
total_answers = total_students_count - pd.isnull(students_df["4.7"].values).sum()
print("Total answers: %s" % (total_answers))
      
unique_answers = [x for x in students_df["4.7"].unique() if not pd.isnull(x)]
all_answers = [x for x in students_df["4.7"] if not pd.isnull(x)]

data = []
for answer in unique_answers:
    count = all_answers.count(answer)
    data.append((count/len(all_answers), answer))
    
student_workplace_choice_df = pd.DataFrame(data, columns=["Percentage", "Label"])
student_workplace_choice_df
```
